載入所需套件

```
install.packages("ggplot2")
install.packages("SportsAnalytics")
```

讀取套件

```
library(SportsAnalytics)
library(ggplot2)
```

利用order()排序球員總分

```
NBA1516<-fetch_NBAPlayerStatistics("15-16")

NBA1516order<-NBA1516[order(NBA1516$TotalPoints,decreasing = T),]
NBA1516order[1:20,]
```

用subset()搜尋火箭隊

```
subset(NBA1516,Team=="HOU")
```

用grepl()列出名字有John的人

```
NBA1516[grepl("John",NBA1516$Name),]
```

散布圖 
x = ThreesAttempted 
y = ThreesMade

```
qplot(ThreesAttempted,ThreesMade,data=NBA1516,color=Position)
```

長條圖 
x = ThreesMade

```
qplot(ThreesMade,data=NBA1516,color=Position)
```

折線圖 
x = Turnovers 
y = Steals


```
ggplot(NBA1516,aes(x=Turnovers,y=Steals))+geom_line()
```

趨勢圖 
x = Turnovers 
y = Steals

```
qplot(Turnovers,Steals,data=NBA1516,geom = c("point","smooth"))
```
