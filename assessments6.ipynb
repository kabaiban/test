```
install.packages("ggplot2")
install.packages("SportsAnalytics")
```

```
library(SportsAnalytics)
library(ggplot2)

NBA1718<-fetch_NBAPlayerStatistics("17-18")
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1718)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1718,color=Position)
```

```
qplot(TotalPoints,data=NBA1718,fill=Position)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1718,facets=Position~.)
```

```
qplot(TotalPoints,data=NBA1718,facets=Position~.,binwidth=100)
```
