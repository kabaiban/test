```
pkgs <- c("tidyverse","lubridate","showtext","readr")
install.packages(pkgs)
install.packages("devtools")
devtools::install_github('cttobin/ggthemr')
#安裝所需套件

library(ggthemr)
library(tidyverse)
library(lubridate)
library(showtext)
library(readr)
#跑所需套件
```

```
showtext.auto(enable = TRUE)
point <- read.csv("682b84de07067bf3770f031e9349e3c4_export.csv")
point
#顯示資料
```

```
point %>% count(縣市別)
#顯示縣市次數總和
```

```
point %>%
  count(縣市別) %>%
  ggplot(aes(x=縣市別,y=n,fill=縣市別))+
  geom_col()+
  geom_text(aes(label=n),vjust=-0.25)+
  labs(title="不同縣市之出遊次數對比",x="縣市",y="數量")+theme(legend.position="none")
  
  point %>%
  count(縣市別) %>%
  ggplot(aes(x = fct_reorder(縣市別,n),y=n,fill=縣市別))+
  geom_col()+
  labs(title="不同縣市之出遊次數對比",x="縣市",y="數量")+
  theme(legend.position="none")
  #各縣市出遊次數長條圖
```

```
point %>%
  count(年別) %>%
  ggplot(aes(x=年別,y=n,fill=年別))+
  geom_col()+
  geom_text(aes(label=n),vjust=-0.25)+
  labs(title="不同年之出遊次數對比",x="年別",y="數量")+theme(legend.position="none")
  
  point %>%
  count(年別) %>%
  ggplot(aes(x = fct_reorder(年別,n),y=n,fill=年別))+
  geom_col()+
  labs(title="不同年之出遊次數對比",x="年別",y="數量")+
  theme(legend.position="none")
  #各年份出遊次數長條圖
```

```
install.packages("ggridges")

library(ggridges)

point %>%
  ggplot(aes(x = 年別,y = 縣市別,fill = 縣市別))+
  geom_density_ridges()
  #各縣市和年份對比圖
```

```
point %>%
  ggplot(aes(x = 年別,y = 類型,fill = 類型))+
  geom_density_ridges()
  #各年份出遊地點類型對比圖呦
```
